```python
total = float(df.shape[0])
ploting = sns.countplot(x='GENDER', data=df)
for p in ploting.patches:
    height = p.get_height()
    ploting.text(p.get_x() + p.get_width()/2.,
           height,
           '{:.2f}'.format((height/total)*100),
           ha='center')
           
plt.legend()
plt.show()
```
